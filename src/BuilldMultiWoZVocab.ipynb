```python
vocab_file = "../resource/woz3/vocab.txt"
feat_file = "../resource/woz3/feat.json"
text_file = "../resource/woz3/text.json"
template_file = "../resource/woz3/template.txt"
```

```python
import json

# read file
with open(feat_file, 'r') as myfile:
    data=myfile.read()
# parse file
obj = json.loads(data)
```

```python
vocab = []
for item in obj.keys():
#     print(item)
    dialogue = obj[item]
    for turn in (dialogue.keys()):
        utt = dialogue[turn]
#         print(utt.keys())
#         print(utt)
        for intent in utt.keys():
            vocab.append(intent)
            temp = utt[intent]
            for value in temp:
                for va in value:
                    vocab.append(va)
       
```

```python
# print(len(vocab))
# vocab_count =  [[x,vocab.count(x)] for x in set(vocab)]
```

```python
from collections import Counter
count = Counter(vocab)
count = sorted(count.items(), key=lambda pair: pair[1], reverse=True)
```

```python
full_vocab = []
mentioned_Set = set()

with open("../resource/woz3/vocab_mod.txt", 'r', encoding='utf8') as vocab_f:
    for line in vocab_f:
        pieces = line.split()
        mentioned_Set.add(pieces[0])
        full_vocab.append(pieces)
```

```python
print(len(mentioned_Set))
```

```python
for item in count:
    if not item[0] in mentioned_Set:
        if (not item[0].__contains__(" ")) and (not item[0].__contains__("\t")):
            full_vocab.append([item[0],item[1]])
    else:
        print(item[0])
```

```python
print(len(full_vocab))
```

```python
for item in full_vocab:
    temp = item[0]
    if temp.__contains__('1'):
        print(temp)
```
